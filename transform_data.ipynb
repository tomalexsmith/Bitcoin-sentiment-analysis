```python
import pandas as pd
```

```python
path = '/content/drive/My Drive/btcTweetWithSentiment.csv'

with open(path, 'r', encoding = 'utf-8') as f:
  tweets_df = pd.read_csv(f, index_col=0)

tweets_df.head(len(tweets_df))
```

```python
tweets_df['mdy'] = pd.to_datetime(tweets_df['mdy'] )
```

```python
daily_average = tweets_df.groupby(tweets_df['mdy'].dt.date)['Sentiment'].mean().reset_index()
```

```python
daily_average.head(len(daily_average))
```

```python
path = '/content/drive/My Drive/avgDailySentiment.csv'

# with open(path, 'w', encoding = 'latin-1') as f:
#   tweets_df.to_csv(f)


with open(path, 'w', encoding = 'utf-8') as f:
  daily_average.to_csv(f)
```

```python
weekly_average.head(len(weekly_average))
```

# Get most popular

```python
tweets_df['mdy'] = pd.to_datetime(tweets_df['mdy'] )
```

```python
tweets_df_10popular = tweets_df.groupby(tweets_df['mdy'].dt.date).apply(lambda x: x.nlargest(10, 'likes'))
```

```python
tweets_df_10popular.head(len(tweets_df_10popular))
```

```python
tweets_df_10popular = tweets_df_10popular.drop('mdy', axis=1)
tweets_df_10popular = tweets_df_10popular.reset_index()
```

```python
tweets_df_10popular = tweets_df_10popular.drop('level_1', axis=1)
```

```python
tweets_df_10popular = tweets_df_10popular.drop('date', axis=1)
```

```python
tweets_df_10popular.head(len(tweets_df_10popular))
```

```python
tweets_df_10popular['mdy'] = pd.to_datetime(tweets_df_10popular['mdy'] )
```

```python
tweets_df_10popular.set_index('mdy', inplace=True)

weekly_average = tweets_df_10popular.resample('W')['Sentiment'].mean().reset_index()
```

```python
weekly_average.head(len(weekly_average))
```

```python
path = '/content/drive/My Drive/TopAvgWeeklySentiment.csv'

# with open(path, 'w', encoding = 'latin-1') as f:
#   tweets_df.to_csv(f)


with open(path, 'w', encoding = 'utf-8') as f:
  weekly_average.to_csv(f)
```
