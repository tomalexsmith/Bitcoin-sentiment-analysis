```python
import snscrape.modules.twitter as snscrape
import pandas as pd
from tqdm.notebook import tqdm
import datetime
```

### Decrement date of tweet scraping

```python
def update_date(curr_day, curr_month, curr_year):

    days_30 = [4, 6, 8, 11]
    prev_day, prev_month, prev_year = curr_day, curr_month, curr_year
    curr_day -= 1

    if curr_month - 1 in days_30:
        num_days = 30
    elif curr_month - 1 == 2:
        if curr_year % 4 == 0:
            num_days = 29
        else:
            num_days = 28
    elif curr_month - 1 == 0:
        num_days = 31
    else:
        num_days = 31
    if curr_day == 0:
        curr_month -= 1
        curr_day = num_days
    if curr_month == 0:
        curr_year -= 1
        curr_month = 12
        curr_day = 31

    return curr_day, curr_month, curr_year, prev_day, prev_month, prev_year
```

### Scrape tweets for each day of specified period

```python
tweets = []
n_tweets_per_day = 100
n_days = 365 # 122, 122, 121
curr_day, curr_month, curr_year = 7, 3, 2023

for day in range(int(n_days)):

    curr_day, curr_month, curr_year, prev_day, prev_month, prev_year = update_date(curr_day, curr_month, curr_year)

    scraper = snscrape.TwitterSearchScraper(
        '(#Bitcoin OR #BTC) min_faves:10 lang:en until:{}-{}-{} since:{}-{}-{}'.format(str(prev_year), str(prev_month).zfill(2),
                                                                          str(prev_day).zfill(2), str(curr_year),
                                                                          str(curr_month).zfill(2),
                                                                          str(curr_day).zfill(2)))

    for i, tweet in tqdm(enumerate(scraper.get_items()), total=n_tweets_per_day - 1):
        data = [
            tweet.date,
            tweet.rawContent,
            tweet.likeCount,
            tweet.retweetCount
        ]
        tweets.append(data)
        if i > n_tweets_per_day - 2:
            break
```

```python
tweets_df = pd.DataFrame(tweets, columns=['date', 'content', 'likes', 'retweets'])
```

```python
tweets_df.dropna(inplace=True)

tweets_df.reset_index(inplace=True, drop=True)
```

```python
# Convert column data types

tweets_df['content'] = tweets_df['content'].astype(str)

tweets_df['likes'] = pd.to_numeric(tweets_df['likes'])
tweets_df['likes'] = tweets_df['likes'].astype(int)

tweets_df['retweets'] = pd.to_numeric(tweets_df['retweets'])
tweets_df['retweets'] = tweets_df['retweets'].astype(int)

tweets_df['date'] = pd.to_datetime(tweets_df['date'], utc=True)
```

```python
tweets_df['mdy'] = tweets_df['date'].dt.date

tweets_df = tweets_df[['mdy','date','content','likes','retweets']]
```
